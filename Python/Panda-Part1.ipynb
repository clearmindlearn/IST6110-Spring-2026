## Create a DataFrame from a dictionary or an excel file

A **DataFrame** is a **2D table** with **rows + columns** (similar to an Excel sheet).

A common way to build a DataFrame:
- Create a **dictionary**
- Each **key** becomes a **column name**
- Each **value** is a **list** (all lists must have the same length)
- You can also create a dataframe from an excel file (e.g., csv excel file)


```python
# Note that data is a dictionary

# - keys   -> column names
# - values -> lists of column values
data = {
    "customer_id": ["C001", "C002", "C003", "C004", "C005", "C006"],
    "customer": ["Alice", "Ben", "Chris", "Diana", "Ethan", "Fiona"],
    "total_order_price": [120.50, 89.99, 45.00, 210.25, 15.75, 62.40],
    "returned": [False, True, False, False, True, False]
}

df1 = pd.DataFrame(data) # DataFrame method converts a dict to a dataframe
df1
```

```python
df1.shape #row x column
```

```python
df1.describe()
```

### Dataframe index

*   In a DataFrame, row labels are called the index
*   Even if it looks like 0,1,2,3 ..., it still a **label**, not a real "data column




```python
df.head()
# pay an attention to the index of the following dataframe:
# index --> 0, 1, 2, 3, 4 ....
# Note that the index is not a real number; it is a label (text)
```

### Row subsetting/slicing

*   .iloc = integer position -based
*   .loc = label based



### iloc = integer position-based

```python
df.iloc[0] #first row by (integer) position --> returns the first row

```

```python
df.iloc[1] # returns the second row ([1] --> 2nd row)
```

```python
#range
df.iloc[0:5] #returns the first 5 rows
#[0:5] --> 0, 1,2,3,4  (note that 5 is excluded)
# 0 -> 1st row, 1 --> 2nd row, 2 --> 3rd row, ....
```

```python
#specific  rows by position
df.iloc[[0,3, 10]] # multple rows --> use double brackets!!
# 0 --> 1st row, 3 --> 4th row, 10 --> 11th row
```

### .loc = label-based (note that default index 0,1,2,3 are labels, not numbers)

*   .loc selects rows by lable (index label)




```python
df.loc[0] # 0 is an index --> 0 is a label; returns the row whose index is 0
```

#### iloc vs loc

```python
df.iloc[0:4] #integer based
#[0:4] --> 0,1,2,3 (note that 4 is excluded)
# 0-> first row, 1 --> 2nd row...
```

```python
df.loc[0:4] # label based; returns all the rows whose indexes are from 0 to 4
```
